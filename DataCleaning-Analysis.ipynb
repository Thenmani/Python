### Inpatient Hospital Data Analysis ###

### Data Loading and Cleaning ###

```python
import pandas as pd

# Loading file data
inpatient_data = pd.read_excel('Inpatient Data.xlsx')

# Displaying first 10 rows
inpatient_data.head(10)
```

```python
# Dataset Shape
inpatient_data.shape
print("Shape of dataset: ", inpatient_data.shape)
```

```python
# Dataset Info
inpatient_data.info()
```

```python
# Standardazing column name as per given meta data
inpatient_data.rename(columns={
    'Patient_ID': 'PatientID',
    'Discarge_status': 'DischargeStatus',
    'name': 'PatientName',
    'sex': 'Sex',
    'age': 'Age',
    'ComorbidityCount': 'ChronicConditionsCount',
    'FamilySupport': 'FamilySupportScore',
    'InsurancePolicy_ID': 'InsurancePolicyID' 
}, inplace=True)

# Verifying column name change
inpatient_data.info()
```

```python
is_missing = inpatient_data.isnull().values.any()
print("Is any values missing? ", is_missing)
```

```python
# No. of missing values per column
missing_per_col = inpatient_data.isnull().sum()
print(missing_per_col)
```

```python
# Replacing missing age with median

# Create a copy for comparison
inpatient_data_original = inpatient_data.copy()

age_median = inpatient_data['Age'].median()
print("Median Age:", age_median)

# Replace age with Median
inpatient_data['Age'] = inpatient_data['Age'].fillna(age_median)

# Verify whether the age is replaced with Median
replaced_age = inpatient_data[inpatient_data_original['Age'].isnull()]
replaced_age.head(2)
```

```python
# To drop rows where “AdmissionSource” is missing.
inpatient_data = inpatient_data.dropna(subset=['AdmissionSource'])
```

```python
#Verification, after dropping rows
# No. of missing values per column
missing_per_col = inpatient_data.isnull().sum()
print(missing_per_col)

print("Shape after cleaning: " ,inpatient_data.shape)
```

```python
# Convert age data type integer safely
inpatient_data['Age'] = inpatient_data['Age'].round().astype('Int64')

# Verify
print(inpatient_data['Age'].dtype)
inpatient_data.head(2)
```

### Statistical Analysis ###

```python
# To calculate average age of patients

average_age = inpatient_data['Age'].mean().round().astype(int)
print("Average age of patients: ", average_age)
```

```python
# Discharge status of each care level

# Grouping by CareLevel
carelevel_group = inpatient_data.groupby('CareLevel')

# Count Discharge Status
dischargestatus_count = carelevel_group['DischargeStatus'].value_counts(sort=False)

# Renaming Discharge Status values
dischargestatus_count = dischargestatus_count.rename(index={0: "Expired", 1: "Discharged alive"}, level='DischargeStatus')

print("Discharge Status grouped by Care Level:\n")
print(dischargestatus_count)

```

```python
# Average hospital bill paid by survivors vs non-survivors

# Grouping by DischargeStatus
discharge_group = inpatient_data.groupby('DischargeStatus')

# Compute average HospitalBill per DischargeStatus
avg_bill_by_status = discharge_group['HospitalBill'].mean()

print("Average hospital bill paid by survivors vs non-survivors:\n")
print(round(avg_bill_by_status,2))
```

```python
# Correlation between hospital bill and patient age using NumPy

import numpy as np

# Extract numeric columns
age = inpatient_data['Age'].to_numpy()
bill = inpatient_data['HospitalBill'].to_numpy()

# Compute Pearson correlation
correlation = np.corrcoef(age, bill)[0, 1]

print("Correlation between Age and HospitalBill:", round(correlation,2))

print("Correlation Matrix:\n")
inpatient_data[['Age', 'HospitalBill']].corr()


```
