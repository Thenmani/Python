```python
import requests

def get_places_by_zip(country_code, zip_code):
    # Construct the URL 
    url = f"https://api.zippopotam.us/{country_code}/{zip_code}"

    try:
        response = requests.get(url)

        # Check if the request was successful (HTTP 200)
        if response.status_code == 200:
            data = response.json()

            print(f"Post Code: {data['post code']}")
            print(f"Country: {data['country']}")

            for place in data['places']:
                print(f" - Place Name: {place['place name']}")
                print(f" - State: {place['state']}")
                print(f" - Longitude: {place['longitude']}")
                print(f" - Latitude: {place['latitude']}")
        else:
            print(f"Error: Received status code {response.status_code}")

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")

# Verification
get_places_by_zip("us", "33131")
```
